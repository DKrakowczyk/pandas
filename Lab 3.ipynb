```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/palles77/Pandas20212022_Lab3/main/employees10000.csv') 
df_Copy = df.copy()
```

```python
#----------ZADANIE 1.1----------
result = df_Copy[df_Copy['Last Name'].str.startswith('A')]['Age in Company (Years)'].mean()
print(f'Pracownicy z nazwiskiem na literę A (średnia): {result}')
```

```python
#----------ZADANIE 1.2----------
result = df_Copy['Salary'].agg({"Srednia pensja":"mean","Odchylenie std. pensji":"std"})
print(f'Odchylenie standardowe i średnia pensja pracownikow: \n {result}')
```

```python
#----------ZADANIE 1.3----------
result = df_Copy[df_Copy['Gender'] == 'F']['Salary'].mean()
print(f'Płaca kobiet (średnia): {result}')
```

```python
#----------ZADANIE 1.4----------
result = df_Copy[df_Copy['Gender'] == 'M']['Salary'].mean()
print(f'Płaca mężczyzn (średnia): {result}')
```

```python
#----------ZADANIE 1.5----------
result = df_Copy[df_Copy['Gender'] == 'M']['First Name'].unique()
print(f'Unikalne imiona męskie:')

for x in result:
    print(x)
```

```python
#----------ZADANIE 1.6----------
result = df_Copy[df_Copy['Gender'] == 'F']['Last Name'].unique()
print(f'Unikalne imiona żeńskie:')

for x in result:
    print(x)
```

```python
#----------ZADANIE 1.7----------
result = df_Copy['City'].unique()
print(f'Unikalne miasta:')
for x in result:
    print(x)
```

```python
#----------ZADANIE 1.8----------
charArr = ('A','B','C','D','E','F','G','H','I','J','K','L','M','N')
result = df_Copy[(df_Copy['Gender'] == 'F') & (df_Copy['First Name'].str.startswith(charArr, na=False))].sort_values('County')['County'].unique()
print(f'Unikalne hrabstwa dla kobiet z imionami z przedzialu A-N:')

for x in result:
    print(x)
```

```python
#----------ZADANIE 1.9----------
min = df_Copy['Salary'].min()
max = df_Copy['Salary'].max()
print(f'Najniższe zarobki: { min }')
print(f'Najwyższe zarobki: { max }')
```

```python
#----------ZADANIE 1.10----------
df_Copy = df.copy()
result = df_Copy[(df_Copy['Gender'] == 'F') & (df_Copy['E Mail'].str.endswith('yahoo.com'))]['E Mail'].count()
print(f'Liczba maili w domenie yahoo.com (kobiety): {result}')
```

```python
#----------ZADANIE 1.11----------
df_Copy = df.copy()
result = df_Copy[(df_Copy['Gender'] == 'M') & (df_Copy['E Mail'].str.endswith('gmail.com'))]['E Mail'].count()
print(f'Liczba maili w domenie gmail.com (mężczyźni): {result}')
```
