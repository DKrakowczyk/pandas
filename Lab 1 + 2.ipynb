```python
import pandas as pd

people = [['Ewa','Kowalska','Warszawa'],["Michał", "Malinowski","Kraków"],["Krzysiek","Kowalski","Gdańsk"],["Kasia","Nowak","Poznań"],["Łucja","Kowalska","Łódź"]]
df_people = pd.DataFrame(people)
df_people.columns = 'Imie', 'Nazwisko', 'Miasto'
df_people
```

Na potrzeby wygenerowania listy 20 osób został stworzony prosty generator losowych danych

```python
from random import randrange
new_people = []
names = ['Agnieszka', 'Basia','Piotrek', 'Zbigniew', "Marcin", 'Adam', 'Weronika']
surnames = ['Wojska', 'Czerniewski','Wojtyła', 'Piela', "Jop", 'Boniek']
cities = ["Bytom", "Sosnowiec", "Gliwice", "Sopot", "Poznań"]
proffesions = ['Malarz', 'Pracownik fizyczny', 'Programista', 'Kierowca']
streets = ['ul. Biała', 'ul. Czarna', 'ul. Beżowa', 'ul. Niebieska']
for x in range(0,20):
 new_people.append([names[randrange(0,len(names))],surnames[randrange(0,len(surnames))],cities[randrange(0,len(cities))],randrange(1960,1996),proffesions[randrange(0,len(proffesions))],streets[randrange(0,len(streets))]])

df = pd.DataFrame(new_people)
df.columns = 'Imie', 'Nazwisko', 'Miasto', "Rok urodzenia", "Zawód", "Adres"
df
```

```python
df.to_csv("people.csv")
```
